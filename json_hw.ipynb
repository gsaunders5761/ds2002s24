```python
import pandas as pd
import json
import requests

#extract json data
url = "https://raw.githubusercontent.com/jasonwnc/ds2002s24/main/2.%20Python/coll%20bb_SortByConf.json"
response = requests.get(url)
json_data = response.text

#load json data
dict = json.loads(json_data)

#remove teams key
teams = dict['teams']

#pandas data frame
df = pd.DataFrame(teams)


# first 100 rows
print("First 100 rows of the DataFrame:")
print(df.head(100))

# a. # of teams
num_teams = df['tid'].nunique()
print("\na. Number of teams:", num_teams)

# b. # of teams in va
va_teams = df['tid'][df['state'] == 'VA']
print("\nb. Number of VA teams:", va_teams.count())

# c. duplicate mascots grouped by count
# find duplicates
duplicate_mascots = df[df.duplicated(subset=['name'], keep=False)]
# count + sort by how many
duplicate_mascots_count = duplicate_mascots.groupby('name').size()
print("\nc. Duplicate Mascots grouped by count:")
print(duplicate_mascots_count.sort_values(ascending=False))
```
